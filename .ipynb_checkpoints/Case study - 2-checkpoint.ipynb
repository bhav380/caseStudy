```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
startup_funding = pd.read_csv('/home/codex/Downloads/startup_funding.csv')
startup_funding.head(50)
```

```python

def correctStartupNames(sf):
    sf.StartupName = sf['StartupName'].str.replace('Oyo Rooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('Oyorooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('OYO Rooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('Paytm Marketplace','Paytm')
    sf.StartupName = sf['StartupName'].str.replace('Ola Cabs','Ola')
    sf.StartupName = sf['StartupName'].str.replace('Olacabs','Ola')
    sf.StartupName = sf['StartupName'].str.replace('OyoRooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('Flipkart.com','Flipkart')
    
    return sf.StartupName


    
```

```python
def top5InvestorsDiffStartups(sf):
    
    sf['InvestorsName'].fillna('unknown',inplace=True)
    sf['InvestorsName'] = sf['InvestorsName'].str.title()
    sf = sf[(sf['InvestorsName']!='Undisclosed Investor') & (sf['InvestorsName']!='Undisclosed Investors')]
    sf.reset_index(drop=True,inplace=True)
    
    
    dic={}
    row=0
    
    
# storing investor name and different companies he has invested on , in dic using set  
#                            structure  ----> dic{'investorname':set('companyA','companyB'...list of companies)}


    for investors in sf['InvestorsName']:

        for investor in investors.split(','):

            if(investor!=''):

                investor = investor.strip();

                setList = dic.get(investor,set())
                setList.add(sf['StartupName'][row])
                dic[investor] = setList;

        row+=1;


    # mainitaing min heap to store top 5 investors 
    #             operation on heap ---> heapq.heappush(topInvestors,(no. of different investments,investorName))

    topInvestors = []
    for investor in dic.keys():

        if(len(topInvestors)<5):
            heapq.heappush(topInvestors,(len(dic.get(investor)),investor))
        else:
            heapq.heappushpop(topInvestors,(len(dic.get(investor)),investor))

    np_topInvestors= np.array(topInvestors)
    
    print(np_topInvestors)

    # sorting topInvestors by number of investments in different companies in decreasing order
    np_topInvestors = np_topInvestors[np_topInvestors[:,0].argsort()[::-1]]


    return np_topInvestors

```




# 1. Most number of fundings  among Bangalore, Mumbai and NCR ( Guragaon, Noida and New Delhi)







```python
sf = startup_funding.copy()

sf['cities'] = sf['CityLocation'].str.title().str.split('/',expand=True)[0]
sf['cities'] = sf['cities'].str.strip()
sf['cities'] = sf['cities'].replace('Delhi','New Delhi')

data = sf['cities'][(sf['cities']=='Bangalore') | (sf['cities']=='Mumbai') | (sf['cities']=='Gurgaon') | (sf['cities']=="Noida") | (sf['cities']=='New Delhi')].str.strip().value_counts()

data['NCR'] = data['Noida'] + data['Gurgaon'] + data['New Delhi']
del data['Noida']
del data['Gurgaon']
del data['New Delhi']

plt.bar(data.index,data)
plt.title('Number of fundings in NCR , Bangalore , Mumbai')
plt.xlabel('Locations')
plt.ylabel('Number of fundings')
plt.yticks(np.arange(0,900,100))
plt.show()

print("NCR has received most number of fundings 709, among the three locations")

```

# 2. Top 5 investors who have invested maximum number of times (including repeat investments)

```python
from operator import itemgetter
sf = startup_funding.copy()

dic = {}




sf['InvestorsName'].fillna('Unknown',inplace=True);

for investors in sf['InvestorsName'].str.split(','):
    for investor in investors:
        investor = investor.strip()
        dic[investor] = dic.get(investor,0)+1

topInvestors = dict(sorted(dic.items(),key=itemgetter(1),reverse=True)[:5])


plt.pie(topInvestors.values(),labels=topInvestors.keys(),autopct="%.2f%%")
plt.title('Percentage of Investments by top 5 investors')
plt.show()

print("Top 5 investors are : ")
for investor in topInvestors.keys():
    print(investor,': ',topInvestors[investor],' investments')
```

# 3. Top 5 investors who have invested in different number of startups

```python
sf = startup_funding.copy()
sf.head(50)

# investors row index save ,ex :- {accel partners : [Rowindexes of accelpartners]}
```

```python
import heapq

sf = startup_funding.copy()
sf.head(50)

# Correcting important startup names
sf.StartupName = correctStartupNames(sf)

# Removing nan values
sf['InvestorsName'].fillna('unknown',inplace=True)

np_topInvestors = top5InvestorsDiffStartups(sf)

# plotting pie chart
plt.pie(np_topInvestors[0:,0],labels=np_topInvestors[0:,1],autopct="%.2f%%")
plt.title('Percentage of Investments by top 5 investors in different companies')
plt.show()

print('Top 5 investors who have invested in different number of startups are : ')
for i in np_topInvestors:
    print(i[1],':',i[0],' startups')

```
