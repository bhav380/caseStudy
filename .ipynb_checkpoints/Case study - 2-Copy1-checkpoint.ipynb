```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

startup_funding = pd.read_csv('/home/codex/Downloads/startup_funding.csv')
```

```python

def correctStartupNames(sf):
    sf.StartupName = sf['StartupName'].str.replace('Oyo Rooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('Oyorooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('OYO Rooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('Paytm Marketplace','Paytm')
    sf.StartupName = sf['StartupName'].str.replace('Ola Cabs','Ola')
    sf.StartupName = sf['StartupName'].str.replace('Olacabs','Ola')
    sf.StartupName = sf['StartupName'].str.replace('OyoRooms','Oyo')
    sf.StartupName = sf['StartupName'].str.replace('Flipkart.com','Flipkart')
    
    return sf.StartupName


    
```

```python
def correctingInvestmentTypeNames(sf):

    sf['InvestmentType'].replace('SeedFunding','Seed Funding',inplace=True)
    sf['InvestmentType'].replace('PrivateEquity','Private Equity',inplace=True)
    sf['InvestmentType'].replace('Crowd funding','Crowd Funding',inplace=True)
```

```python
def top5InvestorsDiffStartups(sf):
    
    # Cleaning Investor Names and removing data of undisclosed Investors from sf
    sf['InvestorsName'].fillna('Undisclosed Investors',inplace=True)
    sf['InvestorsName'] = sf['InvestorsName'].str.title()
    sf['InvestorsName'] = sf['InvestorsName'].str.replace(' & Others','')
    sf['InvestorsName'] = sf['InvestorsName'].str.replace(' & ',',')
    sf['InvestorsName'] = sf['InvestorsName'].str.replace(' And Others','')
    sf['InvestorsName'] = sf['InvestorsName'].str.replace(' And ',',')
    
    sf = sf[(sf['InvestorsName']!='Undisclosed Investor') & (sf['InvestorsName']!='Undisclosed Investors')]
    sf.reset_index(drop=True,inplace=True)
    
    
# storing investor name and different companies he has invested on , in dic using set .
#                            structure  ----> dic{'investorname':set('companyA','companyB'...list of companies),... other entries}

    dic={}
    row=0

    for investors in sf['InvestorsName']:

        for investor in investors.split(','):
         
            if(investor!=''):

                investor = investor.strip();
                setList = dic.get(investor,set())
                setList.add(sf['StartupName'][row])
                dic[investor] = setList;

        row+=1;

    # mainitaing min heap to store top 5 investors 
    #             operation on heap ---> heapq.heappush(topInvestors,(no. of different investments,investorName))

    topInvestors = []
    for investor in dic.keys():

        if(len(topInvestors)<5):
            heapq.heappush(topInvestors,(len(dic.get(investor)),investor))
        else:
            heapq.heappushpop(topInvestors,(len(dic.get(investor)),investor))
            

    np_topInvestors= np.array(topInvestors)    #converting to numpy array
    
    # sorting topInvestors by number of investments in different companies in decreasing order
    np_topInvestors = np_topInvestors[np_topInvestors[:,0].argsort()[::-1]]

    return np_topInvestors
```




# 1. Most number of fundings  among Bangalore, Mumbai and NCR ( Guragaon, Noida and New Delhi)







```python
sf = startup_funding.copy()

cities = sf['CityLocation'].str.title().str.split('/',expand=True)  #splitting cityLocations in each cell and saving saving each city in cities
cities = cities.iloc[0:,0:].replace('Delhi','New Delhi') # replacing Delhi with New Delhi in cities (n X 2 matrix) 

cities.fillna('unknown',inplace=True)
cities[0] = cities[0].str.strip()   #removing extra spaces from start and end of city name
cities[1] = cities[1].str.strip()

data = cities[0].append(cities[1])  #appending col1 to col0 of cities and saving the result inside data 
data.reset_index(drop=True,inplace=True)  #reseting labels of data

#Extracting number of funding in Bangalore,Mumbai and NCR (Gurgaon,Noida,New Delhi)
data= data[(data=='Bangalore') | (data=='Mumbai') | (data=='Gurgaon') | (data=="Noida") | (data=='New Delhi')].value_counts()
data.loc['NCR'] = data['Noida']+data['New Delhi']+data['Gurgaon']
data.drop(data.index[2:5],inplace=True) # dropping data of Noida , New Delhi and Gurgaon

# Plotting pie chart
plt.bar(data.index,data,color=['Purple','Green','Blue'])
plt.title('Number of fundings in NCR , Bangalore , Mumbai')
plt.xlabel('Locations')
plt.ylabel('Number of fundings')
plt.yticks(np.arange(0,900,100))
plt.show()

# printing Result
print("NCR has received most number of fundings 709, among the three locations")



```

# 2. Top 5 investors who have invested maximum number of times (including repeat investments)

```python
from operator import itemgetter
sf = startup_funding.copy()

dic = {}

sf['InvestorsName'].fillna('Undisclosed Investor',inplace=True)


# filling investors and no. of investments made by them in dic
for investors in sf['InvestorsName'].str.split(','):
    for investor in investors:
        investor = investor.strip()
        dic[investor] = dic.get(investor,0)+1
        
        
# finding top 5 Investors who have invested maximum number of times 
topInvestors = dict(sorted(dic.items(),key=itemgetter(1),reverse=True)[:5])

plotting pie chart
plt.pie(topInvestors.values(),labels=topInvestors.keys(),autopct="%.2f%%")
plt.title('Percentage of Investments by top 5 investors')
plt.show()



# printing result
print("Top 5 investors are : ")
for investor in topInvestors.keys():
    print(investor,': ',topInvestors[investor],' investments')
```

# 3. Top 5 investors who have invested in different number of startups

```python
import heapq

sf = startup_funding.copy()

sf.StartupName = correctStartupNames(sf) # Correcting important startup names

np_topInvestors = top5InvestorsDiffStartups(sf)  # fetching top 5 Investors who have invested in different startups

# plotting pie chart
plt.pie(np_topInvestors[0:,0],labels=np_topInvestors[0:,1],autopct="%.2f%%")
plt.title('Percentage of Investments by top 5 investors in different companies')
plt.show()

# printing result
print('Top 5 investors who have invested in different number of startups are : ')
for i in np_topInvestors:
    print(i[1],':',i[0],' startups')

```

# 4. Top 5 investors who have invested in a different number of startups and their investment type is Crowdfunding or Seed Funding.

```python
sf = startup_funding.copy()



correctingInvestmentTypeNames(sf) # Correcting Investment Type

sf = sf[(sf['InvestmentType']=='Seed Funding') | (sf['InvestmentType']=='Crowd Funding')]

sf.StartupName = correctStartupNames(sf) # Correcting important startup names

np_topInvestors = top5InvestorsDiffStartups(sf) # fetching top 5 Investors who have invested in different startups


# plotting pie chart
plt.pie(np_topInvestors[0:,0],labels=np_topInvestors[0:,1],autopct='%.2f%%')
plt.title('Percentage of Investments by top 5 investors in different companies with investment type as Crowd Funding or Seed Funding')
plt.show()


# printing result
print('Top 5 investors with investment type as Crowd Funding or Seed Funding , who have invested in different number of startups are : ')
for i in np_topInvestors:
    print(i[1],':',i[0],' startups')



```

# 5. Top 5 investors who have invested in a different number of startups and their investment type is Private Equity

```python
sf = startup_funding.copy()

correctingInvestmentTypeNames(sf) # Correcting Investment Type Names

sf = sf[(sf['InvestmentType']=='Private Equity')]

sf.StartupName = correctStartupNames(sf) # Correcting important startup names

np_topInvestors = top5InvestorsDiffStartups(sf) # fetching top 5 Investors who have invested in different startups


# plotting pie chart
plt.pie(np_topInvestors[0:,0],labels=np_topInvestors[0:,1],autopct='%.2f%%')
plt.title('Percentage of Investments by top 5 investors in different companies with investment type as Private Equity')
plt.show()

# printing result
print('Top 5 investors with investment type as Private Equity and have invested in different number of startups are : ')
for i in np_topInvestors:
    print(i[1],':',i[0],' startups')
```
